# Центральная предельная теорема

Рассмотрим показательное распределение с такой функцией плотности:

$$f_X(x) = \lambda e^{-\lambda x}, x \geq 0$$

Будем рассматривать с параметром $\lambda = 1$.

https://en.wikipedia.org/wiki/Exponential_distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

### Генерация выборки объёма 1000

```python
lambda_ = 1
exp = sts.expon(scale=1/lambda_)
sample = exp.rvs(1000)
```

### Построение гистограммы выборки и графика теоритеческой плотности распределения

```python
x = np.linspace(0,6,100)
cdf = exp.pdf(x)
plt.plot(x, cdf, label='theoretical PDF')
plt.hist(sample, bins=30, normed=True, label='hist')
plt.ylabel('number of samples, $f(x)$')
plt.xlabel('$x$')
plt.legend(loc='best')
```

### Генерация 1000 выборок объёма n (5, 10, 50) и построение гистограммы распределений их выборочных средних

```python
sample_num = 1000
sample_volume = [5, 10, 50]
```

```python
distribution_samples = np.zeros((len(sample_volume), sample_num))
for i in range(len(sample_volume)):
    for j in range(sample_num):
        temp = np.array(sts.expon().rvs(size=sample_volume[i]))
        temp_mean = temp.mean()
        distribution_samples[i, j] = temp_mean
```

```python
for i in range(len(sample_volume)):
    plt.hist(distribution_samples[i], bins=30, normed=True)
    plt.title('Exponential probability distribution function for n = ' + str(sample_volume[i]))
    plt.xlabel('x')
    plt.ylabel('pdf')
    plt.show()
    
for i in range(len(sample_volume)):
    plt.hist(
        distribution_samples[i],
        bins=30,
        normed=True,
        label='n = ' + str(sample_volume[i])
    )
    plt.title('Exponential probability distribution function')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('pdf')
```

### Вычисление параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних

```python
distribution_sample_mean = 1/lambda_
distribution_sample_variance = np.zeros(len(sample_volume))
for i in range(len(sample_volume)):
    distribution_sample_variance[i] = (1/lambda_**2) / sample_volume[i]
    
print('Математическое ожидание: ' + str(distribution_sample_mean))
for i in range(len(sample_volume)):
    print('Дисперсия для n = ' + 
          str(sample_volume[i]) + ': ' + 
          str(distribution_sample_variance[i]))
```

### Плотности соответствующие нормальным распределениям для каждой гистограммы

```python
for i in range(len(sample_volume)):
    plt.hist(distribution_samples[i], bins=30, normed=True)
    plt.title('Exponential probability distribution function for n = ' + str(sample_volume[i]))
    plt.xlabel('x')
    plt.ylabel('pdf')
    
    mu = distribution_sample_mean
    sigma = np.sqrt(distribution_sample_variance[i])
    normRv = sts.norm(loc=mu, scale=sigma)    
    x = np.linspace(-1,3,100)
    pdf = normRv.pdf(x)
    plt.plot(x, pdf)
    plt.show()
    
for i in range(len(sample_volume)):
    plt.hist(
        distribution_samples[i],
        bins=30,
        normed=True,
        label='n = ' + str(sample_volume[i])
    )
    plt.title('Exponential probability distribution function')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('pdf')
    
    mu = distribution_sample_mean
    sigma = np.sqrt(distribution_sample_variance[i])
    normRv = sts.norm(loc=mu, scale=sigma)    
    x = np.linspace(-1,3,100)
    pdf = normRv.pdf(x)
    plt.plot(x, pdf)
```

### Вывод
Можно заметить, что распределение выборочных средних достаточно хорошо
описывается нормальным распределением, что является утверждением центральной предельной теоремы.
С ростом объема выборки для вычисления выборочного среднего $n$ точность нормальной аппроксимации увеличивается.
