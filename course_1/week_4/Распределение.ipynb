# Распределение хи-квадрат

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
# формирование распределенной величины и выборки размером 1000
k = 3
rv = sts.chi2(k)
iid = rv.rvs(1000)
x = np.linspace(0, 30, 100)
# построения графика плотности распеределени и гистогораммы хи квадрат
plt.plot(x, rv.pdf(x))
plt.hist(iid, density=True)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
# генерация выборок размера 5, 10, 20
size_n = [5, 20, 30]
for n in size_n:
    e = []# пустой массив для записи значения мат. ожидания
    for _ in range(1000):
        e.append(np.mean(rv.rvs(n)))
    plt.subplots(figsize=(8,6))# создает три отдельных графика
    plt.hist(np.array(e), density=True, label='histogram of sample means')#формирование гистограммы распределения выб. ср.
    rv_n = sts.norm(k, np.sqrt(2 * k / n))# нормальное распределение 
    x = np.linspace(0, 7, 100)
    plt.plot(x, rv_n.pdf(x), label='normal distribution')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('n={}'.format(n))
    plt.legend()
    
    
```

# Вывод
Наиболее точно нормальное распределение опистывает распределение выборочных средних
при размере выборки равным 30
