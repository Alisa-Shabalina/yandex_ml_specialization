```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
```

```python
# задаем распределение - хи-квадрат с 2 степенями свободы
chi2_2_rv = sts.chi2(2)
```

```python
# генерируем случайную выборку
smpl = chi2_2_rv.rvs(1000)
```

```python
# рисуем гистограмму
n, bins, patches = plt.hist(smpl, normed=True, label='Histogram')

# рисуем теоретическое распределение
x = np.arange(0,15,0.1)
pdf=chi2_2_rv.pdf(x)
plt.plot(x, pdf, label='PDF')
plt.legend()
plt.xlabel('value')
plt.ylabel('probability density')
plt.title('PDF and histogram for chi2(2)')
```

# n=5

```python
# задаем объем выборки
ssize = 5

#создаем матрицу из 1000 выборок
smpl=np.zeros((1000,ssize))
for q in range(1000): smpl[q,:] = chi2_2_rv.rvs(ssize)

# считаем выборочное среднее
smpl_mean = np.mean(smpl, axis=1)

# строим гистограмму
n, bins, patches = plt.hist(smpl_mean, normed=True, label='Histogram')

# считаем значения нормального распределения в соответствии с ЦПТ
# среднее хи-квадрат(2) =2, дисперсия=4
mu=2.
sigma=math.sqrt(4./ssize)

# диапазон для графика подобрали руками
x = np.arange(0,6,0.1)
clt_norm = sts.norm(loc=mu, scale=sigma)
norm_pdf=clt_norm.pdf(x)
plt.plot(x, norm_pdf, label='CLT')
plt.legend()
plt.xlabel('value')
plt.ylabel('probability density')
plt.title('Histogram and CLT normal distribution for n=' + str(ssize))
```
